### 1. Select all movies in the "Action" genre.

```python
action_movies = df1[df1["Genre"] == "Action"]
display(action_movies)
```

### 2. Select movies released after 2000

```python
movies_after_2000 = df1[df1["Release_Year"] > 2000]
display(movies_after_2000)
```

### 3. Select movies with an IMDb rating greater than 8.5.

```python
imdb_8_5 = df1[df1["IMDb_Rating"] > 8.5]
display(imdb_8_5)
```

### 4. Select only "Movie Name" and "Genre" columns. 

```python
MovieName_Genre = df1[["Movie Name","Genre"]]
display(MovieName_Genre)
```

### 5. Select movies released between 1990 and 2010.

```python
movies_1990_2010 = df1[(df1["Release_Year"] > 1990) & (df1["Release_Year"] < 2010)][["Movie Name", "Release_Year"]]
display(movies_1990_2010)
```

### 8. Find the average IMDb rating of all movies.

```python
imdb_max = np.max(df1["IMDb_Rating"])
highest_rated_movie = df1[df1["IMDb_Rating"] == imdb_max]  # Get the movie(s) with that rating

print(highest_rated_movie)  # Print the details of the highest-rated movie(s)
```

```python
imdb_average = np.mean(df1["IMDb_Rating"])
print(imdb_average)
```

### 9. Find the total box office collection of all movies.

### 11. Rank movies based on IMDb rating.

```python
df1 = df1.sort_values(by="IMDb_Rating", ascending=False)  # Sort by rating
df1["Rank"] = range(1, len(df1) + 1)  # Assign rank
print(df1[["Rank", "Movie Name", "IMDb_Rating"]])  # Show ranked movies
```
